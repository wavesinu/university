### 1.1 Print top 50 most common words (5 pts)

````python
# YOUR CODE MUST BE HERE
import requests
import re
from collections import Counter

url = 'https://www.gutenberg.org/files/2701/old/moby10b.txt'
response = requests.get(url)
text = response.text.lower()

words = re.findall(r'\b[a-z]+\b', text)

word_counts = Counter(words)

common_words = word_counts.most_common(50)
print(common_words)
````

### 1.2 Plot word frequency (5 pts)

- Sort the word frequency in descending order
- Plot the word frequency
- Plot the word frequency in log-log plot.

````python
# YOUR CODE MUST BE HERE
import matplotlib.pyplot as plt
import numpy as np

sorted_counts = sorted(word_counts.values(), reverse=True)

plt.figure(figsize=(10, 6))
plt.plot(sorted_counts)
plt.xlabel('words')
plt.ylabel('frequency')
plt.title('words frequency')
plt.show()

log_counts = np.log10(sorted_counts)
log_ranks = np.log10(range(1, len(sorted_counts) + 1))
plt.figure(figsize=(10, 6))
plt.plot(log_ranks, log_counts)
plt.xlabel('log words')
plt.ylabel('log frequency')
plt.title('word frequency: log-log plot')
plt.show()
````

````python
from bs4 import BeautifulSoup
import requests

url = "https://cs.stanford.edu/research?items_per_page=All&field_faculty_status_value=active"
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

title_elements = soup.find_all('td', class_="views-field views-field-field-faculty-title fac_prof_view_title")
focus_elements = soup.find_all('td', class_="views-field views-field-field-research-focus fac_prof_view_focus")

faculty_titles = [td.text.strip() for td in title_elements]
research_focuses = [td.get_text(separator=', ').strip() for td in focus_elements]
````

### Draw bar charts on research area contributions of Stanford CS faculty

- For each research area, we want to compute how many professors works on that area.
- If one professor works on n research fields, the contribution to one research field is 1/n.
- The colors for professor ranks (assistant, associate, full professors) may be your own choice.
- Your output should be like:



````python
import matplotlib.pyplot as plt
import collections
import pandas as pd

rank_order = ['Assistant Professor', 'Associate Professor', 'Professor']

research_contributions = collections.defaultdict(lambda: collections.defaultdict(int))

for title, focus_str in zip(faculty_titles, research_focuses):
    focuses = focus_str.split(', ')
    for focus in focuses:
        research_contributions[focus][title] += 1/len(focuses)

fields = []
assistant_counts = []
associate_counts = []
professor_counts = []

for field, contributions in research_contributions.items():
    fields.append(field)
    assistant_counts.append(contributions['Assistant Professor'])
    associate_counts.append(contributions['Associate Professor'])
    professor_counts.append(contributions['Professor'])

data = pd.DataFrame({
    'Research Field': fields,
    'Assistant Professor': assistant_counts,
    'Associate Professor': associate_counts,
    'Professor': professor_counts
})

data = data.sort_values('Research Field')

plt.figure(figsize=(15, 8))

plt.bar(data['Research Field'], data['Assistant Professor'], label='assistant professor')
plt.bar(data['Research Field'], data['Associate Professor'], bottom=data['Assistant Professor'],
        label='associate arofessor')
plt.bar(data['Research Field'], data['Professor'],
        bottom=data['Assistant Professor'] + data['Associate Professor'], label='professor')

plt.title('Stanford Univ. faculty research areas')
plt.xlabel('Research Areas')
plt.ylabel('# of faculty members')
plt.xticks(rotation=90)
plt.legend(loc='upper center')
plt.tight_layout()

plt.show()
````
